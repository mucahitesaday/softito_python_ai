### 🔸 Soru 6:
Bir palindrom kontrolü yapın: verilen kelimeyi tersine çevirip orijinaliyle karşılaştırın.
- `"kaçak"` → palindrom mu?
- `"python"` → palindrom mu?

```python
# Cevabınızı buraya yazın:
# İpucu: kelime[::-1] ifadesi kelimeyi tersine çevirir
kelime1 = "kaçak"
kelime2 = "python"


# kelime 1 kontrolü
if kelime1 == kelime1[::-1]:
    print(kelime1, "palindrom")
else:
    print(kelime1, "palindrom değil")

# kelime 2 kontrolü
if kelime2 == kelime2[::-1]:
    print(kelime2, "palindrom")
else:
    print(kelime2, "palindrom değil")
```

## 5️⃣ Döngüler (for / while)

```python
# Örnek: enumerate ve zip
renkler = ["kırmızı", "yeşil", "mavi"]

for i, renk in enumerate(renkler, start=1):
    print(f"{i}. {renk}")

print("---")

isimler = ["Ali", "Ayşe", "Can"]
notlar = [85, 92, 78]

for isim, not_ in zip(isimler, notlar):
    print(f"{isim}: {not_}")
```

## 7️⃣ Sözlükler (Dictionary)
